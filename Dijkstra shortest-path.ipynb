```python
import numpy as np
from heapdict import *
#Read and parse the adjacency list from file and load into an array G
G = []
with open('SPD.txt') as file:
    for line in file:
        strings = line.split()
        tail = int(strings[0])
        tuplist = []
        for tup in strings[1:]:
            tup = [int(char) for char in tup.split(',')]
            tup.append(tail)
            tuplist.append(tup)
        G.append(tuplist)
```

```python
#Quick implementation of breath first search
import queue
n = len(G)
expl_bool = np.zeros(n)
hp = heapdict()

def BFS(G, i):
    global expl_bool, n, hp
    expl_bool[i] = 1
    q = queue.Queue()
    q.put(i)  
    i = 1
    while q.empty != True:
        v = q.get()
        for arc in G[i]:
            expl_bool[arc[0]] = 1
            q.put(arc[0])
```

```python
def main():
    global G
    hp = heapdict()
    n = len(G)
    A = 1e6*np.ones(n)
    A[0] = 0
    hp[1] = 0
    expl_bool = np.zeros(n)
    X = [1]
    
    while len(X) != n:
        #Choose the node to add to X (if not source)
        print(len(X), len(hp))
        print(X)
        if expl_bool[0] == 0:
            node = 1
            expl_bool[0] = 1
        else:
            tup = hp.popitem()
            node = tup[0]
            A[node-1] = int(tup[1])
        #Update the Dijkstra greedy scores
        for edge in G[node-1]:
            w = edge[0]
            if expl_bool[w-1] == 0:
                lvw = edge[1]
                score = A[node-1] + lvw
                try:
                    prev_score = hp[w]
                    if prev_score > score:
                        hp[w] = score
                except KeyError:
                    hp[w] = score
        if node != 1:
            X.append(node)
            expl_bool[node-1] = 1
    return A, X
            
        
```

```python
#Execute Dijkstra algorithm
A, X = main()
```
